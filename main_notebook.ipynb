```python
# Import requests library
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
import getpass
```

```python
apikey = getpass.getpass()
```

```python
#https://www.rijksmuseum.nl/api/nl/collection?key=
```

```python
response = requests.get('https://www.rijksmuseum.nl/api/nl/collection?key='+apikey)
#response
results = response.json()
results
```

```python
results.keys()
```

```python
results['artObjects'][0]['id'] # add the index to get rid of the list
```

```python
results['artObjects'][0]['title'] # add the index to get rid of the list
```

```python
rijks_art_list = pd.DataFrame.from_dict(results, orient='index').reset_index()
rijks_art_list.columns = ["index", "description"]
rijks_art_list
```

```python
rijks_art_objects = rijks_art_list['description'][4]


#rijks_art_objects = pd.DataFrame(rijks_art_objects)

rijks_art_objects
```

```python
flattened_data = pd.json_normalize(rijks_art_objects) # gives the list APIs available
flattened_data
```

```python
facets = flattened_data['facets']
```

```python
flattened_data = pd.json_normalize(facets)
flattened_data
```
